```python
# How do healthcare visits vary for people with chronic vs. non-chronic conditions?
```

```python
chronic_visits = df.groupby(['nchronic', 'lchronic'])['visits'].mean().reset_index()
print(chronic_visits)
plt.figure(figsize=(8, 6))
sns.barplot(x='nchronic', y='visits', hue='lchronic', data=chronic_visits)
plt.title('Average Healthcare Visits Based on Chronic Conditions')
plt.xlabel('Chronic Condition (nchronic)')
plt.ylabel('Average Visits')
plt.show()
```

```python
# Does age or income influence the number of healthcare visits?
```

```python
df.columns
```

```python
chronic_df = df[df['lchronic'] == 'yes']
plt.figure(figsize=(10, 6))

# Plot income vs visits for chronic individuals
sns.scatterplot(x='income', y='visits', data=chronic_df, hue='freepoor', style='freerepat', palette='coolwarm')
plt.title("Income and Free Healthcare Support Impact on Healthcare Visits (Chronic Conditions)")
plt.xlabel("Income")
plt.ylabel("Healthcare Visits")
plt.show()
```

```python
# 4. What is the effect of gender on healthcare utilization among individuals with chronic diseases?
# We will compare healthcare visits by gender for individuals with chronic conditions
chronic_gender = chronic_df.groupby('gender')['visits'].sum().reset_index()

# Plot the result
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='visits', data=chronic_gender)
plt.title("Healthcare Visits by Gender for Chronic Conditions")
plt.xlabel("Gender")
plt.ylabel("Total Healthcare Visits")
plt.show()
```

```python
# Drop the 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Multiply 'age' by 100 and then convert to 'int64'
df['age'] = (df['age'] * 100).astype('int64')

# Round 'income' and convert to 'int64'
df['income'] =(df['income']*100).astype('int64')

# Print the data types to verify the changes
print(df.dtypes)
```

```python
df
```

```python
df.isna().sum()
```

```python
# Healthcare visits for chronic patients by healthcare plan type
chronic_plans = df[df['nchronic'] == 'yes'].groupby(['private', 'freepoor'])['visits'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='private', y='visits', hue='freepoor', data=chronic_plans)
plt.title('Healthcare Visits Based on Healthcare Plans for Chronic Conditions')
plt.xlabel('Private Healthcare Plan')
plt.ylabel('Average Visits')
plt.show()
```

```python
# Healthcare visits by age and income for chronic patients
chronic_df = df[df['nchronic'] == 'yes']
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='income', size='visits', hue='visits', data=chronic_df, palette='viridis', sizes=(20, 200))
plt.title('Healthcare Visits by Age and Income for Chronic Conditions')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(title='Visits')
plt.show()
```

```python
# Compare healthcare visits based on chronic condition status (nchronic)
chronic_visits = df.groupby('nchronic')['visits'].mean().reset_index()
print(chronic_visits)

# Plot the results
plt.figure(figsize=(8, 6))
sns.barplot(x='nchronic', y='visits', data=chronic_visits)
plt.title('Average Healthcare Visits for Chronic vs Non-Chronic Conditions')
plt.xlabel('Chronic Condition (0 = No, 1 = Yes)')
plt.ylabel('Average Number of Visits')
plt.show()
```

```python
# Healthcare visits by gender and chronic condition status

gender_chronic_visits = df.groupby(['gender', 'nchronic'])['visits'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='visits', hue='nchronic', data=gender_chronic_visits)
plt.title('Healthcare Visits by Gender and Chronic Condition Status')
plt.xlabel('Gender')
plt.ylabel('Average Visits')
plt.show()
```

```python
# Heatmap: Chronic Conditions and Visits by Age Group and Gender
df['age_group'] = pd.cut(df['age'], bins=[0, 30, 60, 100], labels=['0-30', '31-60', '61+'])
heatmap_data = df.pivot_table(index='age_group', columns='gender', values='visits', aggfunc='mean')
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar_kws={'label': 'Average Healthcare Visits'})
plt.title('Average Healthcare Visits by Age Group and Gender')
plt.xlabel('Gender')
plt.ylabel('Age Group')
plt.show()
```

```python
# Compare healthcare visits based on chronic condition status (lchronic)
chronic_visits = df.groupby('lchronic')['visits'].mean().reset_index()
print(chronic_visits)

# Plot the results
plt.figure(figsize=(8, 6))
sns.barplot(x='lchronic', y='visits', data=chronic_visits)
plt.title('Average Healthcare Visits for Chronic vs Non-Chronic Conditions')
plt.xlabel('Chronic Condition (0 = No, 1 = Yes)')
plt.ylabel('Average Number of Visits')
plt.show()
```
